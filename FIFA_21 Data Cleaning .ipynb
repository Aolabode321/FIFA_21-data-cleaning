```python
def extract_contract_info(contract):
    if contract == 'Free' or 'On Loan' in contract:
        start_date = np.nan
        end_date = np.nan
        contract_length = 0
    else:
        start_date, end_date = contract.split(' ~ ')
        start_year = int(start_date[:4])
        end_year = int(end_date[:4])
        contract_length = end_year - start_year
    return start_date, end_date, contract_length

# apply fn to contract and create new columns

new_cols = ['Contract Start', 'Contract End', 'Contract Length(years)']
new_data = FIFA_21['Contract'].apply(lambda x: pd.Series(extract_contract_info(x)))

for i in range(len(new_cols)):
    FIFA_21.insert(loc=FIFA_21.columns.get_loc('Contract')+1+i, column=new_cols[i], value=new_data[i])
```

```python
# contract categories

def categorise_contract_status(contract):
    if contract == 'Free':
        return 'Free'
    elif 'On Loan' in contract:
        return 'On Loan'
    else: 
        return 'Contract'
    
# add contract status colums

FIFA_21.insert(FIFA_21.columns.get_loc('Contract Length(years)')+1, 'Contract Status', FIFA_21['Contract'].apply(categorise_contract_status))
```

### 12. Positions

Player's preferred position on the field

```python
FIFA_21['Height'].dtype
```

```python
FIFA_21['Height'].unique()
```

```python
def convert_height(height):
    if "cm" in height:
        return int(height.strip("cm"))
    else:
        feet, inches = height.split("'")
        total_inches = int(feet)*12 + int(inches.strip('"'))
        return round(total_inches*2.54)
    
# apply fn to height column
FIFA_21['Height'] = FIFA_21['Height'].apply(convert_height)
FIFA_21['Height'].unique()
```

```python
# rename Height column
FIFA_21 = FIFA_21.rename(columns={'Height':"Height(cm)"})
FIFA_21.sample(1)
```

```python
def convert_weight(weight):
    if "kg" in weight:
        return int(weight.strip("kg"))
    else:
        pounds = int(weight.strip("lbs"))
        return round(pounds/2.205)
    
# apply fn to weight colunm
FIFA_21['Weight'] = FIFA_21['Weight'].apply(convert_weight)
FIFA_21['Weight'].unique()
```

### 20. Value

Estimated market value of the player.

```python
FIFA_21['Value'].dtype
```

```python
FIFA_21['Value'].unique
```

```python
def convert_value(value):
    if '.' in value:
        million, thousand = value.split('.')
        total_m = int(million.strip('€'))+int(thousand.strip('M'))/10
        return total_m
    elif "M" in value:
        million = int(value.strip('€M'))
        return million
    else:
        euros_k = int(value.strip("€K"))
        return round((euros_k/1000),3)
    
#apply fn to value column
FIFA_21['Value'] = FIFA_21['Value'].apply(convert_value)
FIFA_21['Value'].unique()
```

```python
# rename value to value(€.M)
# rename weight column
FIFA_21 = FIFA_21.rename(columns={"Value":"Value (€.M)"})
```

```python
# fn to convert and make into float

def convert_wage(wage):
    if "K" in wage:
        return int(wage.strip('€K'))
    else:
        wage_conv = int(wage.strip('€'))
        return round((wage_conv/1000),2)
    

#apply fn to wage column
FIFA_21['Wage'] = FIFA_21['Wage'].apply(convert_wage)
FIFA_21['Wage'].unique()
```
